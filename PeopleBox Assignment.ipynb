```python

def calculate_tenure(start_date, end_date):
    if end_date is None:
        end_date = datetime.date.today()
    return (end_date - start_date).days

def convert_data(first_dataset):
    second_dataset = []
    
    for row in first_dataset:
        employee_code = row["Employee Code"]
        manager_employee_code = row["Manager Employee Code"]
        date_of_joining = datetime.datetime.strptime(row["Date of Joining"], "%Y-%m-%d").date()
        date_of_exit = datetime.datetime.strptime(row["Date of Exit"], "%Y-%m-%d").date() if row["Date of Exit"] else None
        last_compensation = row["Compensation"]
        compensation = row["Compensation"]
        last_pay_raise_date = None
        variable_pay = 0
        performance_rating = 0
        engagement_score = 0
        
        for i in range(1, 3):
            if row[f"Compensation {i}"]:
                last_compensation = row[f"Compensation {i}"]
                last_pay_raise_date = datetime.datetime.strptime(row[f"Compensation {i} date"], "%Y-%m-%d").date()
        
        for i in range(1, 3):
            if row[f"Review {i} date"]:
                performance_rating += float(row[f"Review {i}"])
        
        for i in range(1, 3):
            if row[f"Engagement {i} date"]:
                engagement_score += float(row[f"Engagement {i}"])
        
        tenure_in_org = calculate_tenure(date_of_joining, date_of_exit)
        effective_date = date_of_joining
        end_date = date_of_exit if date_of_exit else datetime.date(2100, 1, 1)
        
        second_dataset.append({
            "Employee Code": employee_code,
            "Manager Employee Code": manager_employee_code,
            "Last Compensation": last_compensation,
            "Compensation": compensation,
            "Last Pay Raise Date": last_pay_raise_date,
            "Variable Pay": variable_pay,
            "Tenure in Org": tenure_in_org,
            "Performance Rating": performance_rating,
            "Engagement Score": engagement_score,
            "Effective Date": effective_date,
            "End Date": end_date
        })
    
    return second_dataset





```
